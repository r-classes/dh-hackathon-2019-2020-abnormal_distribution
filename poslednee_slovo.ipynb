```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
poslednee_slovo_html = requests.get('http://lastword.pythonanywhere.com/')
poslednee_slovo_html.encoding = 'utf-8'
poslednee_slovo_html = poslednee_slovo_html.text
poslednee_slovo_html = BeautifulSoup(poslednee_slovo_html, 'html.parser')
```

```python
urls = []
for element in poslednee_slovo_html.find_all('a'):
    urls.append(element.get('href'))
```

```python
urls
```

```python
kto = []
kogda = []
po_kakoj_statje = []
za_chto = []
chto_poluchil = []
for element in urls:
    r = requests.get('http://lastword.pythonanywhere.com{}'.format(element[1:]))
    r.encoding = 'utf-8'
    r = r.text
    r = r.split("<p>")
    regexpwho = '<span>Кто:</span>(.*)[\n]?</p>'
    regexpwhen = '<span>Когда:</span>(.*)[\n]?</p>'
    regexpstatja = '<span>По какой статье:</span>(.*)[\n]?</p>'
    regexpreason = '<span>За что:</span>(.*)[\n]?</p>'
    regexppunishment = '<span>Что получил:</span>(.*)[\n]?</p>'
    for who in r:
        reg = re.findall(regexpwho, who)
        if reg:
            kto.append(reg)
    for when in r:
        reg = re.findall(regexpwhen, when)
        if reg:
            kogda.append(reg)
    for statja in r:
        reg = re.findall(regexpstatja, statja)
        if reg:
            po_kakoj_statje.append(reg)
    for reason in r:
        reg = re.findall(regexpreason, reason)
        if reg:
            za_chto.append(reg)
    for punishment in r:
        reg = re.findall(regexppunishment, punishment)
        if reg:
            chto_poluchil.append(reg)
```

```python
kto_flatten = []
kogda_flatten = []
po_kakoj_statje_flatten = []
za_chto_flatten = []
chto_poluchil_flatten = []

for el in kto:
    kto_flatten.append(el)
for el in kogda:
    kogda_flatten.append(el)
for el in po_kakoj_statje:
    po_kakoj_statje_flatten.append(el)
for el in za_chto:
    za_chto_flatten.append(el)
for el in chto_poluchil:
    chto_poluchil_flatten.append(el)
    
kto_flatten = [y for x in kto_flatten for y in x]
kogda_flatten = [y for x in kogda_flatten for y in x]
po_kakoj_statje_flatten = [y for x in po_kakoj_statje_flatten for y in x]
za_chto_flatten = [y for x in za_chto_flatten for y in x]
chto_poluchil_flatten = [y for x in chto_poluchil_flatten for y in x]
```

```python
vozrast = []
prof = []
regexpvozrast = ".*, .*, (\d*)-го г.р."
regexpprof = ".*, (.*), \d*-го г.р."
regeximya = "(.*), .*, \d*-го г.р."
for el in kto_flatten:
    reg = re.findall(regexpvozrast, el)
    if reg:
        vozrast.append(reg)
    else:
        vozrast.append("н")
for el in kto_flatten:
    reg = re.findall(regexpprof, el)
    if reg:
        prof.append(reg)
    else:
        prof.append("н")
vozrast_flatten = [y for x in vozrast for y in x]
prof_flatten = [y for x in prof for y in x]

print(len(vozrast_flatten))
print(len(prof_flatten))
```

```python
slovo_df = pd.DataFrame()
for i, element in enumerate(urls):
    slovo_df = slovo_df.append(pd.DataFrame(
    [{'Кто':kto_flatten[i],
     'Когда':kogda_flatten[i],
      'По какой статье':po_kakoj_statje_flatten[i],
      'За что':za_chto_flatten[i],
      'Что получил':chto_poluchil_flatten[i],
     'Год рождения':vozrast_flatten[i],
     'Профессия':prof_flatten[i]}]),
    sort = True)
```

```python
slovo_df = slovo_df.reset_index(drop=True)
slovo_df
```

```python
slovo_df.to_csv(path_or_buf="./poslednee_slovo_python_ver.tsv", sep="\t")
```
